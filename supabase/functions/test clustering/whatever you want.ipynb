```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
```

```python
print(f"Number of samples: {len(newsgroups.data)}")
print(f"Number of categories: {len(newsgroups.target_names)}")
print(f"Categories: {newsgroups.target_names}")
```

```python
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(newsgroups.data)
```

```python
print(X.shape)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())
```

```python
print(X_pca.shape)
```

```python
kmeans = KMeans(n_clusters=20, random_state=42)
kmeans.fit(X)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('K-means Clustering of 20 Newsgroups Dataset')
plt.show()
```

```python
for i in range(20):
    print(f"Cluster {i}:")
    cluster_docs = [newsgroups.data[j] for j in range(len(newsgroups.data)) if kmeans.labels_[j] == i]
    #print(cluster_docs[:5])  # Print first 5 documents in each cluster
    print(len(cluster_docs))
    print("\n")
```
